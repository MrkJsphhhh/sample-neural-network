```python
#Import the packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras import Sequential
from keras.layers import Dense
```

```python
#Read .csv file
titanic = pd.read_csv('Testing.csv')
```

```python
#Data Cleaning
titanic.isna()
```

```python
titanic.isna().sum()
```

```python
titanic = titanic.dropna(subset = ['Age', 'Cabin', 'Embarked'])
titanic = titanic.drop(columns = ['Name', 'Cabin', 'Ticket'])
```

```python
titanic.isna().sum()
titanic.head()
```

```python
#Feature Classification 
titanic['Sex'].unique()
```

```python
titanic['Embarked'].unique()
```

```python
sex = {'male': 1,'female': 0}
embark = {'S': 0,'Q': 1, 'C': 2}
```

```python
titanic['Sex'].replace(sex, inplace=True)
titanic['Embarked'].replace(embark, inplace=True)
```

```python
titanic.head()
```

```python
# X and y representation
X = titanic.drop(columns = ['Survived'])
y = titanic ['Survived']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
X_train.shape
X_test.shape
#y_train.shape
#y_test.shape
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#num_epochs = 20
#history=model.fit(padded, training_tags_final, epochs=num_epochs, validation_data=(testing_padded, testing_tags_final))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Fit the model on the dataset
model.fit(X, y, epochs=100, batch_size=30)
#Evaluate the model
_,accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
model.summary()
```
